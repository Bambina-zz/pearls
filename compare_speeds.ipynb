```
## Import any libarary
import matplotlib.pyplot as plt
import pandas as pd

## Import data 
df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/compare_speeds.csv')
df.index = range(1, df['juggling'].count()+1)

plt.figure()
df.plot()
plt.xlabel("Rotation distance")
plt.ylabel("Speed(sec)")
plt.xlim(0, df['juggling'].count()+1)
plt.show()
```
